```python
counts=0
countns=0
plt.title('Ratio of Survivers vs Non Survivers')
for i in range(len(train)):
    if train.loc[i,'Survived']==1:
        counts+=1
    else:
        countns+=1
plt.pie([counts,countns],colors=['green','red'],labels=['Survived ('+str(counts)+')','Not survived ('+str(countns)+')'],shadow=True,radius=1)
plt.show()
    
```
